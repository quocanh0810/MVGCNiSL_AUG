```python
import pandas as pd

# Đọc file SL_data_original.csv từ đường dẫn
file_path = "SL_data_original.csv"
df = pd.read_csv(file_path)

# Lấy danh sách các gene1 và gene2
genes = pd.concat([df['gene1'], df['gene2']])

# Lọc ra các gene duy nhất
unique_genes = genes.unique()

# Lưu các gene duy nhất vào file SL_unique_MVGCNiSL.txt
output_file = "SL_unique_MVGCNiSL.txt"
with open(output_file, "w") as f:
    for gene in unique_genes:
        f.write(f"{gene}\n")

# Hiển thị kết quả file đã lưu
output_file
```

```python
import pandas as pd

# Đọc file entities.tsv và id_mapping.csv
entities_file = 'TransEdata/kg_embed/entities.tsv'  # Đường dẫn tới file entities.tsv
id_mapping_file = 'TransEdata/kg_embed/id_mapping.csv'  # Đường dẫn tới file id_mapping.csv

# Đọc file entities.tsv, file này không có tiêu đề
entities_df = pd.read_csv(entities_file, sep='\t', header=None, names=['id', 'name'])
id_mapping_df = pd.read_csv(id_mapping_file)

# Tạo một từ điển ánh xạ id -> name từ file id_mapping.csv
id_to_name_map = dict(zip(id_mapping_df['id'], id_mapping_df['name']))

# Ánh xạ id với tên tương ứng trong entities.tsv
entities_df['name_mapped'] = entities_df['id'].map(id_to_name_map)

# Lưu file kết quả với tên entities_name.tsv
output_file = 'TransEdata/kg_embed/entities_name.tsv'
entities_df.to_csv(output_file, sep='\t', index=False, header=True)

print(f"File entities_name.tsv đã được tạo và lưu tại: {output_file}")
```

```python
#Bước 7:

import numpy as np
from sklearn.decomposition import PCA

# Bước 1: Đọc file .npy
npy_file = './TransEdata/Jurkat/common_genes_embedding.npy'
embedding_data = np.load(npy_file)

# Bước 2: Thực hiện PCA để giảm số chiều từ 400 xuống 128
pca = PCA(n_components=128)
reduced_embedding_data = pca.fit_transform(embedding_data)

# Bước 3: Kiểm tra kích thước mới sau khi giảm chiều
print("Kích thước sau khi giảm chiều:", reduced_embedding_data.shape)

# Bước 4: Lưu kết quả vào file .npy mới
np.save('./TransEdata/Jurkat/common_genes_embedding_128.npy', reduced_embedding_data)
print("File common_genes_embedding_128.npy đã được tạo.")

# Kiểm tra bao nhiêu phần trăm thông tin được giữ lại
explained_variance = pca.explained_variance_ratio_
print("Tỉ lệ variance được giữ lại:", sum(explained_variance))

print(reduced_embedding_data)
```

```python
#Bước 10:
import pandas as pd
import numpy as np

# Đọc file CSV chứa danh sách gene và id
csv_file_path = "../data/TransEdata/Jurkat/common_genes_Jurkat_MVGCNiSL_embeding_with_id.csv"
df = pd.read_csv(csv_file_path)

# Đọc file embedding chứa embedding của 407 gene
embedding_file_path = "../data/TransEdata/Jurkat/common_genes_embedding_128.npy"
embeddings = np.load(embedding_file_path)

# Sắp xếp lại file CSV theo cột unique_genes (giữ lại cột id cũ để đối chiếu sau)
df_sorted = df.sort_values(by='unique_genes').reset_index(drop=True)

# Cột 'id' cũ thể hiện vị trí ban đầu của gene trước khi sắp xếp
old_ids = df_sorted['id'].values

# Dùng old_ids để tạo một mảng embedding mới theo thứ tự mới
new_embeddings = embeddings[old_ids]

# Cập nhật lại cột id mới theo thứ tự mới (bắt đầu từ 0)
df_sorted['id'] = range(len(df_sorted))

# Lưu lại file CSV đã sắp xếp
output_csv_path = "../data/TransEdata/Jurkat/common_genes_Jurkat_MVGCNiSL_embeding_sorted_with_id.csv"
df_sorted.to_csv(output_csv_path, index=False)

# Lưu lại file embedding đã sắp xếp
output_embedding_path = "../data/TransEdata/Jurkat/common_genes_embedding_128_sorted.npy"
np.save(output_embedding_path, new_embeddings)

print(f"File CSV đã sắp xếp được lưu tại: {output_csv_path}")
print(f"File embedding đã sắp xếp được lưu tại: {output_embedding_path}")
```

```python
import numpy as np
import pandas as pd

# Đọc file CSV cũ và CSV mới
old_csv_path = "../data/TransEdata/Jurkat/common_genes_Jurkat_MVGCNiSL_embeding_with_id.csv"
new_csv_path = "../data/TransEdata/Jurkat/common_genes_Jurkat_MVGCNiSL_embeding_sorted_with_id.csv"
df_old = pd.read_csv(old_csv_path)
df_new = pd.read_csv(new_csv_path)

# Đọc file embedding cũ và embedding mới
old_embedding_path = "../data/TransEdata/Jurkat/common_genes_embedding_128.npy"
new_embedding_path = "../data/TransEdata/Jurkat/common_genes_embedding_128_sorted.npy"
old_embeddings = np.load(old_embedding_path)
new_embeddings = np.load(new_embedding_path)

# Kiểm tra thứ tự gene giữa file cũ và file mới
old_gene_order = df_old['unique_genes'].values
new_gene_order = df_new['unique_genes'].values

# So sánh thứ tự gene giữa hai file
if np.array_equal(np.sort(old_gene_order), new_gene_order):
    print("Thứ tự gene đã được sắp xếp đúng theo ABC.")
else:
    print("Thứ tự gene không khớp!")

# Kiểm tra xem embedding có được sắp xếp đúng không
# So sánh embedding của gene đầu tiên sau khi sắp xếp
for i, gene in enumerate(new_gene_order):
    old_idx = df_old[df_old['unique_genes'] == gene].index[0]  # Lấy vị trí của gene trong file CSV cũ
    if not np.array_equal(new_embeddings[i], old_embeddings[old_idx]):
        print(f"Embedding của gene {gene} không khớp tại vị trí {i} (old index {old_idx})!")
        print(f"Embedding cũ: {old_embeddings[old_idx][:5]}...")  # In ra một phần của embedding cũ
        print(f"Embedding mới: {new_embeddings[i][:5]}...")  # In ra một phần của embedding mới
    else:
        print(f"Embedding của gene {gene} khớp.")
```
